## Write a python program to scrape mentioned news details from https://www.cnbc.com/world/?region=world:-Headline, Time, Hyperlink

```python
# First we will import all the libraries required for web-scraping
# We will require two libraries for web-scraping:
   # 1. requests: This will be used to send get request to the web page server to get the source-code for the webpage
        
    #2. BeautifulSoup: It will be used to prase the source code and to extract the required data from the parsed structure.
```

### Import required libraries

```python
# importing the required libraries

import requests
from bs4 import BeautifulSoup
import re
import pandas as pd
import numpy as np
```

```python
page = requests.get("https://www.cnbc.com/world/?region=world")
page
```

```python
soup=BeautifulSoup(page.content)
soup
```

```python
headline=[]
```

```python
for i in soup.select(".LatestNews-headline"):
    headline.append(i.text)
headline
```

```python
time=[]
```

```python
for i in soup.select(".LatestNews-timestamp"):
    time.append(i.text)
time
    
```

```python
news_link=[]
```

```python
for link in soup.find_all('a',class_="LatestNews-headline"):
    news_link.append(link['href'])
news_link
```

```python
print(len(headline),len(time),len(news_link))
```

```python

df= pd.DataFrame({'headline':headline, 'time':time, 'news_link':news_link})
```

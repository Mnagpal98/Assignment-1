#### Write a python program to scrape the details of most downloaded articles from AI in last 90 days.
https://www.journals.elsevier.com/artificial-intelligence/most-downloaded-articles Scrape below mentioned details: Paper title,Authors,Publish Date, Paper url

```python
# First we will import all the libraries required for web-scraping
# We will require two libraries for web-scraping:
   # 1. requests: This will be used to send get request to the web page server to get the source-code for the webpage
        
    #2. BeautifulSoup: It will be used to prase the source code and to extract the required data from the parsed structure.
```

### Import required libraries

```python
# importing the required libraries

import requests
from bs4 import BeautifulSoup
import re
import pandas as pd
import numpy as np
```

```python
page = requests.get("https://www.journals.elsevier.com/artificial-intelligence/most-downloaded-articles")
page
```

```python
soup=BeautifulSoup(page.content)
soup
```

```python
paper_title=[]
```

```python
for i in soup.select('h2'):
    paper_title.append(i.text)
paper_title
```

```python
authors=[]
```

```python
for i in soup.find_all('span',"sc-1w3fpd7-0 dnCnAO"):
    authors.append(i.text)
authors
```

```python
published_date=[]
```

```python
for i in soup.find_all('span',class_="sc-1thf9ly-2 dvggWt"):
    published_date.append(i.text)
published_date
```

```python
paper_url=[]
```

```python
for link in soup.find_all('a',class_="sc-5smygv-0 fIXTHm"):
    paper_url.append(link['href'])
paper_url
```

```python
df= pd.DataFrame({'paper_title':paper_title,'Authors':authors,'Published_date':published_date,'Paper_url':paper_url})
```
